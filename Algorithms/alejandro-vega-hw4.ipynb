## Problem 1. [30]
Give a **recursive implementation** of a dynamic programming algorithm for the Best Steal problem  
that we discussed in class. The input will be two arrays v and s (value and size) and the  
capacity C. Your algorithm should return the **list of items** that should be stolen in order to  
maximize the value, without exceeding the capacity C.

```python
from pprint import pprint
bookKeep = []

def backItUp(i,j, items, s):
    global bookKeep
    # if we have no remaining capacity or we have no items remaining
    if (j == 0 or i < 0):
        return items
    # if our index pointer is different than our current capacity
    # and the current item fits in our "bag" then add to our "bag"
    # keep looking through the items to see what fits in our "bag"
    # with the remaining capacity specified in our book keeping 
    # i.e. we used this item. look through the rest
    elif (j != bookKeep[i][j][1] and j >= s[i]):
        items.append(i)
        return backItUp(i-1,bookKeep[i][j][1],items,s)
    # keep looking through the rest of our items with the same capacity 
    # i.e. we didn't use this item. look through the rest
    else:
        return backItUp(i-1,j,items,s)
        
    
    

def BSR(v, s, i, j):
    global bookKeep
    # reached last item and item i doesn't fit in our bag
    # return 0 since we didn't gain any value
    if(i == 0 and j < s[i]):
        return 0
    # reached last item and item i fits in our bag.
    # return the value of item i since we can "steal" it
    elif (i == 0 and j >= s[i]):
        return v[0]
    # item i doesn't fit in our "bag". skip item i
    elif(j < s[i]):
        return BSR(v,s,i-1,j)
    # skipping item i and "stealing" item i are both valid solutions
    # solve both and store the largest value of the two and the
    # remaing capacity in our book keeping 
    # i.e. decide if we should "steal" item i or skip it
    else:
        used = v[i] + BSR(v, s, i-1, j-s[i])
        unused = BSR(v,s,i-1,j)
        if (used > unused):
            bookKeep[i][j] = (used,j-s[i])
            return used
        else:
            bookKeep[i][j] = (unused, j)
            return unused
    

def bestStealRec(v, s, C):
    # nothing fits in our "bag". return empty list
    if all(i > C for i in s):
        return []
    else:
        global bookKeep
        # initialize our book keeping to 0 value and 
        # a non-sensical remaining capacity
        for j in range(0,len(v)):
            bookKeep.append([(0,-1)]*(C+1))
        BSR(v, s, len(v)-1, C)
        pprint(bookKeep)
        return backItUp(len(v)-1, C, [], s)
        
v = [100,70,40,50]
s = [4,3,2,1]
C = 5
print(bestStealRec(v,s,C))

bookKeep = []

v= [1,4,2,9]
s = [3,2,3,2]
C = 7
print(bestStealRec(v,s,C))

bookKeep = []

v = [1,4,2,9]
s = [3,2,3,2]
C = 6
print(bestStealRec(v,s,C))
```

## Problem 2. [30]
Give an **iterative implementation** of a dynamic programming algorithm for the Best Steal  
problem that we discussed in class. The input will be two arrays v and s (value and size) and  
the capacity C. Your algorithm should return the **list of items** that should be stolen in order to  
maximize the value, without exceeding the capacity C.

```python
def bestStealIter(v,s,C):
    # nothing fits in our "bag". return empty list
    if all(i > C for i in s):
        return []
    
    else:
        b = []
        # initialize our book keeping, b, to 0 value and 
        # a non-sensical remaining capacity
        for n in range(0,len(v)):
            b.append([(0,-1)] * (C+1))
        
        # for the first item check with which capacities, c,  can we "steal" it
        # and save in our book keeping
        for c in range(0,C+1):
            if (c >= s[0]):
                b[0][c] = (v[0], c - s[0])
                
        # for the rest of the items, check for each item,i, for each capacity,j,
        # 1) if the item it too big to steal with our current capacity
        # then the solution is the same as our previous item
        # 2) if it fits:
        # a) ignore it
        # b) steal it
        # check which option leads to the most value 
        # and save the optimal solution in our book keeping
        for i in range(1,len(v)):
            for j in range(0,C+1):
                if(j < s[i]):
                    b[i][j] = b[i-1][j]
                else:
                    used = v[i] + b[i-1][j-s[i]][0]
                    unused = b[i-1][j][0]
                    if (used > unused and j >= s[i]):
                        b[i][j] = (used, j-s[i])
                    else:
                        b[i][j] = (unused, j)
        
        stole = [] # indexes of items that maximize value
        i = len(b)-1
        j = C
        
        # while there's still capacity left 
        # or we haven't checked all items:
        # check if we obtained value from another capacity (column)
        # if so: 
        # 1) we used the current item i
        # 2) search in the column saved in our book keeping
        # else:
        # we didn't use this item. check the next item with the same capacity
        while (j != 0 and i >= 0):
            point = b[i][j][1]
            if (j != point and j >= s[i]):
                stole.append(i)
                i -= 1
                j = point
            else:
                i -= 1         
        return stole
                
v = [100,70,40,50]
s = [4,3,2,1]
C = 5
print(bestStealIter(v,s,C))
        
v= [1,4,2,9]
s = [3,2,3,2]
C = 7
print(bestStealIter(v,s,C))

v = [1,4,2,9]
s = [3,2,3,2]
C = 6
print(bestStealIter(v,s,C))        
```

### Subproblems
We can approach this problem by checking all subsequences smaller than j that end in the ith item.  


Let:  
- our function be $jS$  
- our current last index in $A$ be $i$  
- the size of $A$ be $n$  
- our current size of subsequence be $c$  
- our book keeping, $bk$, be a matrix of size $n$ x $j$ initialized to 0's  
    except the first column which will be initialized with 1's.   
    


We will solve subproblems of jS(n, j) by:  
- (In this implementations) start searching with $i = 2$, $c = 2$ in $bk$ with $jS(i,c)$ ,since  
    this is the first valid incresing subsequence.  
- for all $1 < r < i$, check if $A_i > A_r$ then add the solutions in $bk_{r,c-1}$ to $bk_{i,c}$  
- If $i < c$, then this is not a valid solution. Check next $i$  
- Do this by traversing our matrix, $bk$, until $i = n$ and $c = j$  
- Our final solution will be the sum of the solutions   
    found in the last column of our book keeping matrix.   
    

If we were to solve this problem by solving all smaller subproblems we would:  
- Start serching with $i = n$, $c = j$ in $jS(i,c)$  
- for all $1 < r < i$, $1 < c$ and $c \leq r$ add the solutions in $bk_{r,c-1}$ to $bk_{i,c}$  
- Do this by backtracking through our matrix, $bk$, until $i = 1$ and $c = 1$  
- Our final solution will be the sum of the solutions   
    found in the last column of our book keeping matrix.  




```python
def jLS(A, j, i, c, bk):
    # let n be the size of A
    if (i == len(A)-1 and c == j-1): # if on last row and column
        
        # if previous numbers in A are smaller than A_n where n is the last index
        # add their soulutions to bk[n][j] where j is the last column
        for k in range(i-1, -1, -1): 
            if (A[i] > A[k]):
                bk[i][c] += bk[k][c-1]       
        sum = 0
        for k in range(0, len(A)): # sum all the solutions of subsequences of size j
            sum += bk[k][c]
        return sum
    
    # if i > n, start at row 2 and the next column
    elif (i > (len(A)-1)): 
        return jLS(A, j, 1, c+1, bk)
    
    # if the current column is larger than the current index
    # we can't search for a solution of the subsequence 
    # since there's not enough numbers up to index i
    elif (c > i):
        return jLS(A, j, i+1, c, bk)
    
    # for every number that is smaller than A[i] 
    # add their solutions to bk[i][c]
    else:
        for k in range(i-1, -1, -1):
            if (A[i] > A[k]):
                bk[i][c] += bk[k][c-1]
        return jLS(A, j, i+1, c, bk)
    

def jLongSub(A, j):
    # if looking for subsequences of size 1, there are no increasing subsequences
    if (j == 1):
        return 0
    # not enough numbers for a subsequence of size j
    elif (len(A) < j):
        return 0
    # if exactly j numbers in A, A must be sorted so there can be 1 solution
    elif (len(A) == j):
        if all(A[i] < A[i+1] for i in range(0, len(A)-1)):
            return 1
        else:
            return 0
    # solve all solve problems
    else:
        bk = []
        # initialize our book keeping with 0's except 
        for i in range(0,len(A)):
            bk.append([1] + [0]*(j-1))
        return jLS(A, j, 1, 1, bk)
    
A = [3,2,4,5]
j = 3
print(jLongSub(A, j))

A = [1,3,2,4,6,5]
j = 4
print(jLongSub(A, j))

A = [1,3,2,9,10]
j = 3
print(jLongSub(A, j))

A = [1,1,2,3,4]
j = 3
print(jLongSub(A, j))
```
