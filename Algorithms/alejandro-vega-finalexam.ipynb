### Problem 1. 
Imagine a group of *n* participants in the first round of a chess tournament. Suppose *n* is an  
even number. The tournament committee wants to set up a pairing for the first round. Prove by induction that the   number of pairings among which the committee has to choose one, is equal to the product of the odd numbers from  
1 to n, i.e.  
  
$$(n−1)∗(n−3)∗(n−5)...3∗1$$
  
For example, assuming we have n = 4 persons, A, B, C, D there are 3*1 pairings.  
Pairing 1: {(A,B),(C,D)}  
Pairing 2: {(A,C), (B,D)}  
Pairing 3: {(A,D), (B,C)}  

### Proof:
Before proving this, we must find an explicit formula for all possible unique pairings.  
In general we could do this by choosing the first pair with ${{n}\choose{2}}$, ${{n - 2}\choose{2}}$ for the next pair, and so on up to ${{2}\choose{2}}$.   

Note that:  
  
$${{n}\choose{2}} = \frac{n!}{2!(n-2)!} = \frac{((n-2)+1) \cdot ((n-2)+2)}{2!}$$
  
This means that every ${{n}\choose{2}}$ gives us a numerator $(n) \cdot (n-1)$ and we decrease n in each term by a factor of 2 each time. Note that this gives us $\frac{n}{2}$ terms. Therefore, we have:  
  
$$\frac{n!}{2^{\frac{n}{2}}}$$

Since we're dealing with chess pairings we don't want duplicates (i.e. (ab)(cd) == (cd)(ab)), and we need to get rid of these duplicates. We do this by dividing each term ${{n}\choose{2}}$ by $\frac{n}{2}$. Remember we do this for each term, therefore we end up with:

$$(\frac{n}{2}) \cdot {(\frac{n}{2}-1)} \cdots (\frac{6}{2}) \cdot (\frac{4}{2}) \cdot (\frac{2}{2}) = (\frac{n}{2}) \cdot {(\frac{n-2}{2})} \cdots 3 \cdot 2 \cdot 1 = (\frac{n}{2})!$$
  
This gives us the following explicit formula:

$$\frac{n!}{(\frac{n}{2})! \ \ 2^{\frac{n}{2}}}$$

Disclosure: This [link](https://math.stackexchange.com/questions/1234696/number-of-ways-you-can-form-pairs-with-a-group-of-people-when-certain-people-can) and a very nice Combinatorics student helped me see how we get this formula.

**Prop**: $n \in \mathbb{N},\ n\ even \Rightarrow (n-1)(n-3)\cdots(3)(1) = \frac{n!}{(\frac{n}{2})!2^{\frac{n}{2}}}$  
**Proof (Induction)**: Let $A = \{x \in \mathbb{N},\ x\ even\ |\ (x-1)\cdot(x-3)\cdots(3)\cdot(1) = \frac{x!}{(\frac{x}{2})! \ \ 2^{\frac{x}{2}}}\}$  
  
**Base Case**: Let $k \in \mathbb{N},\ k = 2$ 
  
$(2 - 1)$  
$1 = \frac{2!}{(\frac{2}{2})! \ \ 2^{\frac{2}{2}}} = \frac{2}{1!2^{1}} = 1$  
  
$∴\ k \in A$
  
**Inductive Hypothesis**: $k \in A \Rightarrow (k+2) \in A$  
  
**Inductive Step**:  
$k \in A$  
  
$\Rightarrow (k-1)\cdot(k-3)\cdots(3)\cdot(1)\ =\ \frac{k!}{(\frac{k}{2})! \ \ 2^{\frac{k}{2}}}$  
  
$\Rightarrow (\frac{k+2}{k+2})\cdot((k+2)-1)\cdot(k-1)\cdot(k-3)\cdots(3)\cdot(1)\ =\ \frac{k!}{(\frac{k}{2})! \ \ 2^{\frac{k}{2}}}\cdot(\frac{k+2}{k+2})\cdot((k+2)-1)$  
  
$\Rightarrow (1)\cdot(k+1)\cdot(k-1)\cdot(k-3)\cdots(3)\cdot(1)\ =\ \frac{k!}{(\frac{k}{2})! \ \ 2^{\frac{k}{2}}}\cdot(\frac{k+2}{2(\frac{k}{2}+1)})\cdot(k+1)$  
  
$\Rightarrow (k+1)\cdot(k-1)\cdot(k-3)\cdots(3)\cdot(1)\ =\ \frac{k!}{(\frac{k}{2})! \ \ 2^{\frac{k}{2}}}\cdot(\frac{(k+1)(k+2)}{(\frac{k}{2}+1)2})$  
  
$\Rightarrow (k+1)\cdot(k-1)\cdot(k-3)\cdots(3)\cdot(1)\ =\ \frac{(k+2)!}{(\frac{k}{2}+1)! \ \ 2^{\frac{k}{2}+1}}$  
  
$\Rightarrow ((k+2)-1)\cdot(k-1)\cdot(k-3)\cdots(3)\cdot(1)\ =\ \frac{(k+2)!}{(\frac{k+2}{2})! \ \ 2^{\frac{k+2}{2}}}$ 
  
$\Rightarrow (k+2) \in A$  
  
∴ By the Principle of Mathematical Induction, $k \in A \Rightarrow (k+2) \in A\  \forall k \in \mathbb{N},\ k\ even$

### Problem 2
Write code that takes as input an even number n and prints out all possible pairings of n items  
(for the pairings you can use letters as above, or numbers). Hint: When you see induction, think recursion  
(but not necessarily).  

```python
#!/usr/bin/env python3 

# if we do pairings by hand we notice 
# that in every pair except the last one
# we can fix the first element of the pair,
# assign the ith element of A to the second 
# element, and calculate the next pair with
# the rest of the list we didn't use. 

# function will take a list of "players" A,
# the number of players n, the current list 
# of matchups M, and the list of lists of 
# matchups P (pairings). Function is void and
# instead modifies given P by reference
def pairs(A, n, M, P):
    
	# there's only 1 possible matchup remaining
    # add it to our list of matchups and add
    # this list to our possible matchups for 
    # n players
	if (n == 2):
		M.append((A[0], A[1]))
		P.append(M)
		M = []
	else:
		# when there's more than 1 pair remaining, 
		# we'll fix the first element of the pair 
		# to be the first element of our remaining 
		# players and we'll have n - 1 possible options 
		# for the second element. we'll iterate through
		# these options
        
		for i in range(1,n):
            # create matchup between the 
            # first element of A and the ith
		    matchup = [(A[0], A[i])]
		    # python makes copies by reference 
		    # so we use a slice, but take the
		    # whole list
		    rest = A[:]
		    # remove the elements we just used 
		    # since they're no longer options
		    rest.remove(A[0])
		    rest.remove(A[i])
		    # Add the matchup to our list of 
		    # matchups and find the next pair
		    pairs(rest, n-2, M+matchup, P)

# we'll initialize the parameters for 
# pairs() and return all possible pairings
def pairsWrap(n):
	P = []
    # create a list of numbers 65 to 65+n
    # and map the char cast function to each number
    # to create a list of letters
	players = list(map(chr, range(65, 65+n)))
	pairs(players, n, [], P)
	return P

n = int(input("Enter an even number of players: "))
Pairings = pairsWrap(n)
print()
for i, pairing in enumerate(Pairings):
	formatted = '{' + str(pairing)[1:-1] + '}'
	print("Pairing {}: {}".format(i+1,formatted))
```
